Histograph - used to represent to show frequency distributions


```python
import matplotlib.pyplot as plt
import numpy as np
data=[10,15,20,30,35,57,42,14,52,43,52,53,35,13,53,24,64,24,6,76]
plt.hist(data,bins=20)

plt.show()
```

```python
#optional paremeter ledgent
z=np.linspace(0,1,100)
x=np.sin(z)
y=np.cos(z)
fig=plt.figure(figsize=(15,15))
ax=fig=fig.add_subplot(projection="3d")
ax.scatter(x,y,z,color="green",label="line graph")
ax.set_title("3d line graph")
ax.set_xlabel("x-axis")
ax.set_ylabel("y-axis")

ax.set_zlabel("z-axis")

ax.legend(fontsize=20)
plt.show()
```

from data frame make a graph
2D graph

```python
import matplotlib.pyplot as plt
import pandas as pd
datas={
    "height":[150,160,140,165,133,154,145,167],
    "weight":[50,53,48,59,44,56,51,66],
    "age":[20,22,20,24,19,21,20,25]
}
df=pd.DataFrame(datas)
print(df)
plt.figure(figsize=(9,6))
plt.scatter(df["height"],df["weight"],s=50,color="blue",marker='o')
fig=plt.figure
print(fig)
```

from data frame make a graph
3D graph

```python

datas={
    "height":[150,160,140,165,133,154,145,167],
    "weight":[50,53,48,59,44,56,51,66],
    "age":[20,22,20,24,19,21,20,25]
}
df=pd.DataFrame(datas)
print(df)
fig=plt.figure(figsize=(10,10))
ax=fig.add_subplot(projection="3d")
plt.plot(df["height"],df["weight"],df["age"],color="blue",marker='o',label="line graph")
ax.set_title("3d line graph")
ax.set_xlabel("height")
ax.set_ylabel("weight")
ax.set_label("age")
ax.legend(loc="upper left",fontsize=20)
plt.show()
```

```python
#stack plot

days=[1,2,3,4,5]
sleeping=[6,7,8,9,5]
eating=[2,3,3,4,1]

working=[7,6,8,6,7]
playing=[4,5,4,6,4]
plt.stackplot(days,sleeping,eating,working,playing,
              labels=['sleep','eat','work','play'],
              colors=['blue','c','r','k'])
plt.legend(loc="upper left")
plt.title("daily routine")
plt.xlabel("days")
plt.ylabel("hours")
plt.xticks(days)
plt.show()
```

```python
#box plot(whisker plot) usen statistical  work based in quartile q1,q2,q3 ,lower bound and upper bound
#outlier -out of range data or different type of data from all other data
data=[1,5,6,7,86,7,9,10,50]
plt.boxplot(data,
            notch=True,#creates a notch box
            vert=False,#true create a vertical box if false create a horizontal box
            patch_artist=True,#box are filled with colour
            showmeans=True,#show the median bar pointer
            boxprops=dict(facecolor="lightyellow",color="green"),#determined which color is used by the boxes
            medianprops=dict(color='red',linewidth=2),#show the median bar with red color
            whiskerprops=dict(color='lightyellow',linewidth=1.5),#define the line from q1 to small values and q3 to large values
            capprops=dict(color='purple',linewidth=1.5),#define the external line of the whister plot
            flierprops=dict(marker='o',color='orange',markersize=6)# define the marker of the out line
            )
plt.title("customized box plot")
plt.show()
```

```python
#heat map find relation on the basis of color
import numpy as np
data=np.random.rand(6,6)
plt.imshow(data,cmap="plasma",interpolation="nearest")#imshow() display the data as an image 2d arrat and cmap hot red
#colormap to map data values to colors, 'hot' goes from black -> red-> -> yello->white (heat colors)
# hoe to render pixels between data points, "nearesr" means no smoothing , pixels are shown as blocks
plt.colorbar() # ass color dcale
plt.title("heatmap graph")
plt.show()
```

```python
#task -create a stack diagram of two bar with legend
# for x -axis categories
quarters=['Q0','Q1','Q2','Q3','Q4','Q5','Q6','Q7','Q8']

#height of bar for a
product_a=[20,35,40,15,10,33,21,33,12]


# height of bar for b
product_b=[15,25,35,30,24,45,32,43,43]

#plot the bars for product A
plt.bar(
    quarters,
    product_a,
    label="product_A",
    color="orange",
    width=0.5,
    edgecolor='black'
)


#plot the bars for product B
plt.bar(
    quarters,
    product_b,
    bottom=product_a,
    label="product_B",
    color="purple",
    width=0.5,
    edgecolor='orange',
    alpha=0.5
)

plt.title('quaterly sales', fontsize=15)

plt.ylabel("sales",fontsize=13)
plt.xlabel('quartile',fontsize=13)
plt.legend(loc='upper right')

plt.show()
```
